### import necessary libraries 

```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from autogluon.tabular import TabularDataset, TabularPredictor 
```

### read your file 

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv') 
df.head()
```

```python
df['stroke'].unique()
```

```python
df.describe()
```

```python
df.dtypes
```

### divide your data object to interest parts 

```python
df_train,df_test=train_test_split(df,test_size=0.2,random_state=1)
df_train.shape,df_test.shape
```

```python
test_data=df_test.drop(['stroke'],axis=1)
test_data.head()
```

### create model object 

```python
predictor= TabularPredictor(label ='stroke').fit(train_data = df_train, verbosity = 2,presets='best_quality')
```

### get summarize of your model 

```python
predictor.fit_summary()
```

```python
predictor.leaderboard(df_train, silent=True)
```

```python
predictor.leaderboard(df_train, silent=False)
```

```python
predictor.evaluate(df_train)
```

### show the feature importance 

```python
predictor.feature_importance(data=df_train)
```

### try predictiong  to new data 

```python
y_pred = predictor.predict(test_data)
y_pred=pd.DataFrame(y_pred,columns=['stroke'])
y_pred #print the DataFrame
```

```python
predictor.evaluate(df_test)
```
